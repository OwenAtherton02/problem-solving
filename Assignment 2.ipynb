```python
import numpy as np
import random
import matplotlib.pyplot as plt

max_n = 14
nvals = np.arange(2,max_n+1)
average_wins = []
for n in nvals:
    all_wins = []
    for _ in range(int(1e2)):
        wins = 0
        for _ in range(int(1e3)):
            M = np.zeros((n,n))
            ones = 0
            while ones != n**2//2:
                M[random.randint(0,n-1),random.randint(0,n-1)] = 1
                ones = np.sum(M)
            if np.linalg.det(M) == 0.0:
                wins += 1
        all_wins.append(wins)
    print("n = " + str(n) + " simulated.")
    average_wins.append(np.mean(np.array(all_wins)))

print("Finished all simulations")
```

```python
plt.plot(nvals, average_wins, color = 'purple')
plt.xlabel("n")
plt.ylabel("Wins/1000 (percentage number of matrices with zero-determinant, multiplied by 10)", fontsize = 6)
plt.title("Wins against n")
plt.show()
```
