```python
import numpy as np
from random import randint
from numba import njit
import seaborn as sns
import matplotlib.pyplot as plt


@njit(fastmath = True)
def hat_pull(N):
    hat = [num for num in range(1, N+1)]

    for _ in range(N-1):
        i = randint(0,len(hat)-1)
        a = hat[i]
        hat.remove(a) #doesn't matter if it removes another element that's equal to a
        i = randint(0,len(hat)-1)
        b = hat[i]
        hat.remove(b)
        c = abs(a-b)
        hat.append(c)

    return hat

max_val = 2024

n = hat_pull(max_val)[0]
print (n)

final_papers = []
for _ in range(int(1e6)):
    n = hat_pull(max_val)[0]
    final_papers.append(n)

print(final_papers[0])
print(max(final_papers))
print(min(final_papers))

fig = sns.histplot(final_papers, bins=100, stat = 'count', color = 'Teal')
plt.ylabel("Count")
plt.xlabel("n values")
plt.title("Distibution of n values")
plt.grid()
plt.show()
```

```python
@njit(fastmath = True)
def three_hat_pull(N):
    hat = [num for num in range(1, N+1)]

    for _ in range(N-2):
        i = randint(0,len(hat)-1)
        a = hat[i]
        hat.remove(a) #doesn't matter if it removes another element that's equal to a
        i = randint(0,len(hat)-1)
        b = hat[i]
        hat.remove(b)
        i = randint(0,len(hat)-1)
        c = hat[i]
        hat.remove(c)
        d = abs(a-b)
        e = abs(a-c)
        f = abs(b-c)
        g = max(d,e,f)
        d_e_f = [d,e,f]
        d_e_f.remove(g)
        for i in d_e_f:
            hat.append(i)

    return hat

max_val = 2024

xvals = []
yvals = []
for _ in range(int(1e4)):
    n = three_hat_pull(max_val)
    xvals.append(n[0])
    yvals.append(n[1])

plt.plot(xvals,yvals, 'cx')
plt.show()
```
