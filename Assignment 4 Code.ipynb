```python
import numpy as np
from math import comb
import matplotlib.pyplot as plt
```

```python
T = 24

L = 25 #here L is the number we want n_t to be

prob_of_L = 0
for t in range(1, T+1):
    num_of_combs = comb(t,24-t) if 24-t >= 0 else 0
    total_chains = 2**t
    prob_nt_is_L = num_of_combs / total_chains
    prob_of_L += prob_nt_is_L
    
print (prob_of_L)
```

```python
max_L = 40

L_vals = np.arange(2,max_L+1) #here L_vals are the numbers we want n_t to be. Starting from 2 makes more sense as for L=1 we're starting on 1 and would be rolling 0 times.

P_hat_L_vals = []
for L in L_vals:
    P_hat_L = 0
    for t in range(L//2, L):
        num = comb(t,L-t-1)
        denom = 2**t
        prob = num / denom
        P_hat_L += prob
    P_hat_L_vals.append(P_hat_L)
    
plt.plot(L_vals, P_hat_L_vals, color = 'Purple')
plt.xlabel('$L$')
plt.ylabel('$\hat{P}(L)$')
plt.grid()
plt.show()
```

```python
float_eps = np.finfo(float).eps

L = 1
P_hat_L = 0
while abs(P_hat_L - 2/3) > float_eps:
    L += 1
    P_hat_L = 0
    for t in range(L//2, L):
        num = comb(t,L-t-1)
        denom = 2**t
        prob = num / denom
        P_hat_L += prob
    print(P_hat_L)
    if L>(1e5):
        break
print (L)
```
